```python
import pandas as pd
```

```python
data = pd.read_excel('homicidios.xlsx', sheet_name=None)
```

```python
data_hechos = data['HECHOS']
data_victimas = data['VICTIMAS']
```

```python
data_hechos.head(3)
```

```python
# Drop GeoPandas info not used

data_hechos.drop('XY (CABA)', axis=1, inplace=True)
```

```python
# Show rows completley unusable

data_hechos[data_hechos.isnull().all(axis=1)]
```

```python
# Show rows where at least one value is NaN

data_hechos[data_hechos.isnull().any(axis=1)]
```

```python
# Show rows where at least one of 'Altura' or 'Cruce' value is NaN

data_hechos[['Altura', 'Cruce']][data_hechos[['Altura', 'Cruce']].isnull().any(axis=1)]
```

```python
# Remaining columns with NaN after excluding 'Altura' and 'Cruce'

columns_to_exclude = ['Altura', 'Cruce']
df_except_specific_column = data_hechos.drop(columns=columns_to_exclude)

df_except_specific_column.count()
```

```python
# Delete rows were no location is given
rows_to_delete = df_except_specific_column[df_except_specific_column.isnull().any(axis=1)].index

data_hechos.drop(index=rows_to_delete, inplace=True)
data_hechos
```

```python
# Showing duplicated rows

data_hechos[data_hechos.duplicated()]
```

```python
data_hechos['HORA'].describe()
```

```python
data_hechos[data_hechos['HORA']=='SD']
```

```python
# Store row 518 and later drop it from data df
row_no_hora = data_hechos[data_hechos['HORA']=='SD']
row_no_hora
```

```python
# Drop no location available row

data_hechos.drop(518, inplace=True)
```

```python
data_hechos['FECHA'].dt.date
```

```python
# Concatenate 'FECHA' (date) and 'HORA' (time) columns
data_hechos['FECHA-HORA'] = data_hechos['FECHA'].dt.strftime('%Y-%m-%d') + ' ' + data_hechos['HORA'].astype(str)
data_hechos['FECHA-HORA']
```

```python
# Replace 'SD' with NaN in the entire DataFrame

data_hechos.replace('SD', pd.NA, inplace=True)
data_victimas.replace('SD', pd.NA, inplace=True)
```

```python
data_hechos[['pos x', 'pos y']].info()
```

```python
# Rename 'pos x' to 'long' and 'pos y' to 'lat'

data_hechos.rename(columns={'pos x': 'long'}, inplace=True)
data_hechos.rename(columns={'pos y': 'lat'}, inplace=True)
```

```python
# Convert datatypes to numeric

data_hechos['long'] = pd.to_numeric(data_hechos['long'], errors='coerce')
data_hechos['lat'] = pd.to_numeric(data_hechos['lat'], errors='coerce')
```

```python
# There are 4 rows with no position to plot

data_hechos.loc[data_hechos['lat'].isna() | data_hechos['long'].isna()][['lat', 'long']]
```

```python
# Calculate the mean values of the columns excluding NaN values

mean_value_lat = data_hechos['lat'].mean()
mean_value_long = data_hechos['long'].mean()

# Fill NaN values in both columns with their respective means (for later plotted maps wont diverge from center)

data_hechos['lat'].fillna(mean_value_lat, inplace=True)
data_hechos['long'].fillna(mean_value_long, inplace=True)
```

```python
# Check result

data_hechos.loc[data_hechos['lat'].isna() | data_hechos['long'].isna()][['lat', 'long']]
```
